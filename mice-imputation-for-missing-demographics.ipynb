```python
import pandas as pd
import numpy as np
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import LabelEncoder
```

```python
# Sample Data: 1.0 = Female, 0.0 = Male (Simplified for MICE)
# MICE: Multiple Imputation by Chained Equations --> this looks at other variables (e.g., type of health service accessed) to predict the missing value.
# Service: 0 = General, 1 = Antenatal, 2 = GBV Support
data = pd.DataFrame({
    'age': [24, np.nan, 31, 19, np.nan, 45],
    'gender_encoded': [1.0, 1.0, 0.0, 1.0, np.nan, 0.0],
    'service_type': [1, 1, 0, 2, 1, 0] 
})
print (data)
```

```python
# Initialize the MICE Imputer
mice_imputer = IterativeImputer(max_iter=10, random_state=42)
print (mice_imputer)
```

```python
# Fit and Transform
imputed_data = mice_imputer.fit_transform(data)
print (imputed_data)
```

```python
# Convert back to DataFrame
df_final = pd.DataFrame(imputed_data, columns=data.columns)
print (df_final)
```

```python
# Rounding age and gender back to logical integers
df_final['age'] = df_final['age'].round(0)
# df_final['age'] = data.age
df_final['gender_encoded'] = df_final['gender_encoded'].round(0)
```
